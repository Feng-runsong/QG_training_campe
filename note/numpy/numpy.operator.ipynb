```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
```

* 向量四则运算

```python
print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
print(arr1 / arr2)
```

* 矩阵乘法（两种方式等价）

```python
print(np.dot(arr1, arr2))
print(arr1 @ arr2)
```

* 其它运算

```python
print(arr1 ** 2) # 乘方
print(np.sqrt(arr1)) # 开方
```

* 转置（两种方式等价）

```python
mat = np.array([[1, 2, 3], [4, 5, 6]])
print(mat, end="\n------------------------\n")

print(np.transpose(mat), end="\n------------------------\n")
print(mat.T)
```

* 范数

```python
vec = np.array([3, 4])
print(np.linalg.norm(vec, ord=1))	# L1
print(np.linalg.norm(vec))			  # L2
print(np.linalg.norm(vec, ord=3))	# L3
```

* 相关性分析

```python
np.corrcoef(arr1, arr2) # 皮尔逊系数矩阵
np.cov(arr1, arr2)		  # 协方差矩阵
```
