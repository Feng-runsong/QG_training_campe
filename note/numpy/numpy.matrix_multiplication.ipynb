```python
import time
import numpy as np
```

纯 Python 的矩阵乘法

```python

def pure_python_matrix_multiply(A, B):
    result = []
    for _ in range(len(A)):
        row = []
        for _ in range(len(B[0])):
            row.append(0)
        result.append(row)
    # result = [[0 for _ in range(len(B[0]))] for _ in range(len(A))] # python 列表生成器 简化写法
    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result
```

下面配置测试矩阵的**尺寸**。注意：矩阵中的**每个元素（实数、float）都会占用内存空间**。请不要无节制地扩大矩阵尺寸，并时刻检测你的内存占用变化  
>在老一些的 windows 中，操作系统内存占满后可能导致系统崩溃

准备两样：
- N：矩阵尺寸
- A B 和 A_np B_np：纯 Python 二维数组（列表）和 NumPy 矩阵

```python
# 测试矩阵大小
N = 200
# 生成随机矩阵
A = [[np.random.rand() for _ in range(N)] for _ in range(N)]
B = [[np.random.rand() for _ in range(N)] for _ in range(N)]
# 转换为 NumPy 数组
A_np = np.array(A)
B_np = np.array(B)
```

测试纯 Python 的矩阵乘法时间

```python
start_time = time.time()
python_result = pure_python_matrix_multiply(A, B)
print(f"纯 Python 实现耗时: {(time.time() - start_time):.6f} 秒")
```

测试 NumPy 的矩阵乘法时间

```python
# 测试 NumPy 的矩阵乘法时间
start_time = time.time()
numpy_result = np.dot(A_np, B_np)
print(f"NumPy 耗时: {(time.time() - start_time):.6f} 秒")
```
