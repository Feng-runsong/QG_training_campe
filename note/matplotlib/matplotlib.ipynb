```python
import matplotlib.pyplot as plt

# 示例数据：一个月内每天的销售量
days = list(range(1, 31))  # 1到30天
sales = [250, 300, 280, 310, 290, 330, 340, 360, 300, 310,
         280, 350, 360, 390, 370, 400, 420, 430, 410, 390,
         400, 420, 430, 440, 450, 460, 480, 500, 490, 520]

# 创建图形对象
plt.figure(figsize=(10, 6))

# 绘制柱状图
plt.bar(days, sales, color='lightblue', alpha=0.7, label='Daily Sales')

# 绘制折线图
plt.plot(days, sales, color='blue', marker='o', linestyle='-', label='Trend')

# 添加标题和标签
plt.title('Daily Sales for a Month')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.legend()

# 显示图形
plt.show()
```

* 坑点：`plt.figure`会返回一个`Figure`对象，该对象被创建在了`plt`的全局，如果直接以`plt.func`的形式进行操作，其默认会作用在该全局对象上，显式的面向对象形式操作如下：

```python
# 创建一个 Figure 对象
fig = plt.figure()

# 在 Figure 中添加一个 Axes 对象
ax = fig.add_subplot(111)  # 第一个1代表行数，第二个1代表列数，第三个1代表这是第一个子图

# 创建数据
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# 使用 Axes 对象的 plot 方法绘制线图
ax.plot(x, y)

# 设置标题和轴标签
ax.set_title('Sample Plot')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# 显示图形
plt.show()
```

* 坑点：当需要使用`plt`创建两张不同的图像时，需要分别执行两次`plt.figure()`操作，且在每次执行完`plt.figure()`时会创建 一个新的`Figure`对象，且之后的操作会作用在最近一次创建的`Figure`对象上，具体示例如下：

```python
days = list(range(1, 31))
sales = [250, 300, 280, 310, 290, 330, 340, 360, 300, 310,
         280, 350, 360, 390, 370, 400, 420, 430, 410, 390,
         400, 420, 430, 440, 450, 460, 480, 500, 490, 520]

# 绘制第一个图
plt.figure(figsize=(10, 6))
plt.bar(days, sales, color='lightblue', alpha=0.7, label='Daily Sales')
plt.title('bar chart')
plt.show()

# 绘制第二个图
plt.figure(figsize=(10, 6))
plt.plot(days, sales, color='blue', marker='o', linestyle='-', label='Trend')
plt.title('line chart')
plt.show()
```
